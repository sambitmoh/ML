```
# Write your code here
from sklearn.model_selection import train_test_split
from sklearn import tree
import numpy as np



features = list(Market1.drop(['resp'],1).columns)
X_train= np.array(Market1[features])
y_train =np.array(Market1["resp"])

X_train, X_test, y_train, y_test = train_test_split(X_train ,y_train , train_size = 0.8,random_state = 42)
print("Train_data", X_train.shape)
print("Test_variable", X_test.shape)
print("Train_data", y_train.shape)
print("Test_variable", y_test.shape)
```

```
# Write your code here
DTree = tree.DecisionTreeClassifier( max_depth=4)
DTree.fit(X_train,y_train)
```

```
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree, export_text
plt.figure(figsize=(15,7))
plot_tree(DTree, filled=True, 
                     rounded=True,
                     impurity=False,
                     feature_names = features)
print( export_text(DTree, feature_names = features))
```

### Train-Test; Accuracy and Confusion Matrix

```
## Make predictions
predict1 = DTree.predict(X_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_train, predict1)
print (cm)

total = sum(sum(cm))
#####from confusion matrix calculate accuracy
accuracy = (cm[0,0]+cm[1,1])/total
print(accuracy)
```

```
predict1 = DTree.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predict1)
print (cm)

total = sum(sum(cm))
#####from confusion matrix calculate accuracy
accuracy = (cm[0,0]+cm[1,1])/total
print(accuracy)
```

```
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold
from sklearn.tree import DecisionTreeClassifier
```

```
Grid_Search = {'max_depth' : [2,3,5,10,15,20],'max_leaf_nodes' : [5,7,10,15,20,25,30,35,40]}

```

```
clf = GridSearchCV(DecisionTreeClassifier(),Grid_Search, refit =True)
clf.fit(X_train,y_train)

clf_train_score = clf.score(X_test,y_test)
clf_test_score = clf.score(X_train, y_train)

print("Train score : ", clf_train_score)
print("Test score : ", clf_test_score)
print("best tree: ", clf.best_estimator_)
```

```
plt.figure()
plot_tree(clf.best_estimator_, filled=True, 
                     rounded=True,
                     impurity=False,
                     feature_names = features)
print( export_text(clf.best_estimator_, feature_names = features))
```

```
Train_predict = clf.predict(X_train)
test_predict = clf.predict(X_test)
print("CM1", confusion_matrix(y_train,Train_predict))
print("CM2",confusion_matrix(y_test,test_predict))
```

```
## Classification report on training data
from sklearn.metrics import classification_report
print(classification_report(y_train,Train_predict))
```

```
## Test Confusion Matrix

confusion_matrix(y_test,test_predict)
```

```
## Test Classification Report
print(classification_report(y_test,test_predict))
```
